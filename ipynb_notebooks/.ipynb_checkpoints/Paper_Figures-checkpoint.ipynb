# Location and Home Vocabulary

```python
def plot_clock(axes,hourly_rate,color_data=None,cbarticks=[0.02,0.08]):
    def hour_format(x, p):
        hour = x * 12 / np.pi
        return '{:0.0f}'.format(hour) if x > 0 else '24'
    
    def plot(ax, theta, counts, formatter,color_data):
        if color_data is None:
            cmap = matplotlib.cm.get_cmap('Blues')
            colors=[cmap((x-min(counts))/(max(counts)-min(counts))) for x in counts]
            norm = mpl.colors.Normalize(vmin=min(counts), vmax=max(counts))
            sm = plt.cm.ScalarMappable(cmap=cmap, norm= mpl.colors.Normalize(vmin=min(counts),
                                                                             vmax=max(counts)))
            sm._A = []
            label="Rate"
        else:
            color_data_real=[x for x in color_data if x!=0]
            cmap = matplotlib.cm.get_cmap('Spectral')
            print("Running with variable income. Please change max and min values here if other SES")
            min_real,max_real=9000,28000
            colors=[cmap((x-min_real)/(max_real-min_real))
                    if x !=0 else (0.0,0.0,0.0,1)
                    for x in map(int,color_data)]    
            sm = plt.cm.ScalarMappable(cmap=cmap, norm=mpl.colors.Normalize(vmin=min(color_data_real),
                                                                            vmax=max(color_data_real)))
            label="SES"
            sm._A = []
        
        ax.bar(theta, counts, width=np.pi/12, color=colors, alpha=0.8)
        cb=plt.colorbar(sm,ax=ax,orientation="horizontal",fraction=0.05,
                     pad=0.15,aspect=20,label=label,ticks=cbarticks)
        cb.set_label(label=label,fontsize=12)
        ax.xaxis.set_major_formatter(tkr.FuncFormatter(formatter))
    bins = np.r_[0.5:12, 12.5:24,]
    axes.set(theta_offset=np.pi/2, theta_direction=-1,
           xticks=np.arange(0, np.pi*2, np.pi/12),
           yticks=[round(x,2) for x in np.linspace(0,0.8*hourly_rate.max(),3)[1:]])
    plot(axes,( bins * np.pi / 12), hourly_rate, hour_format,color_data)
```

```python
# Location Data
base_fpr_location_xgb = np.linspace(0, 1, 101)
tprs_location_xgb=[]
for fpr,tpr in zip(dic_location_xgb["fpr"],dic_location_xgb["tpr"]):
    tpr = interp(base_fpr_location_xgb, fpr,tpr)
    tpr[0] = 0.0
    tprs_location_xgb.append(tpr)

tprs_location_xgb = np.array(tprs_location_xgb)
mean_tprs_location_xgb = tprs_location_xgb.mean(axis=0)
std_location_xgb = tprs_location_xgb.std(axis=0)
sqrt_n=1
tprs_upper_location_xgb = np.minimum(mean_tprs_location_xgb + std_location_xgb/sqrt_n, 1)
tprs_lower_location_xgb = np.maximum(mean_tprs_location_xgb - std_location_xgb/sqrt_n, 0)

# Linkedin Data
base_fpr_linkedin_xgb = np.linspace(0, 1, 101)
tprs_linkedin_xgb=[]
for fpr,tpr in zip(dic_linkedin_xgb["fpr"],dic_linkedin_xgb["tpr"]):
    tpr = interp(base_fpr_linkedin_xgb, fpr,tpr)
    tpr[0] = 0.0
    tprs_linkedin_xgb.append(tpr)

tprs_linkedin_xgb = np.array(tprs_linkedin_xgb)
mean_tprs_linkedin_xgb = tprs_linkedin_xgb.mean(axis=0)
sqrt_n=1#
std_linkedin_xgb = tprs_linkedin_xgb.std(axis=0)
tprs_upper_linkedin_xgb = np.minimum(mean_tprs_linkedin_xgb + std_linkedin_xgb/sqrt_n, 1)
tprs_lower_linkedin_xgb = np.maximum(mean_tprs_linkedin_xgb - std_linkedin_xgb/sqrt_n, 0)


# Architect Data
base_fpr_archi_xgb = np.linspace(0, 1, 101)
tprs_archi_xgb,aucs_archi_xgb=[],[]
i=0
for fpr,tpr,auc in zip(dic_archi_xgb["fpr"],dic_archi_xgb["tpr"],dic_archi_xgb["auc"]):
    if i==0 or i==3:
        pass
    else:
        tpr = interp(base_fpr_archi_xgb, fpr,tpr)
        tpr[0] = 0.0
        tprs_archi_xgb.append(tpr)
        aucs_archi_xgb.append(auc)
    i+=1  

tprs_archi_xgb = np.array(tprs_archi_xgb)
mean_tprs_archi_xgb = tprs_archi_xgb.mean(axis=0)
std_archi_xgb = tprs_archi_xgb.std(axis=0)
sqrt_n=1
tprs_upper_archi_xgb = np.minimum(mean_tprs_archi_xgb + std_archi_xgb/sqrt_n, 1)
tprs_lower_archi_xgb = np.maximum(mean_tprs_archi_xgb - std_archi_xgb/sqrt_n, 0)
```
