```python
from tqdm import tqdm
from unidecode import unidecode
```

```python
import re
import pandas as pd
data_fr=pd.read_csv("/warehouse/COMPLEXNET/jlevyabi/ml_soc_econ/icdm18/issues/data_fr_site.csv")
data_en=pd.read_csv("/warehouse/COMPLEXNET/jlevyabi/ml_soc_econ/icdm18/issues/data_en_site.csv")
all_prods=[]
dic_crawled_job_to_insee={}
for it,row in  data_en.iterrows():
    if type(row.job_name) is float:
        continue
    dic_crawled_job_to_insee.setdefault(unidecode(row.job_name.lower()),
                                        (data_fr.iloc[it].insee_job,
                                        data_fr.iloc[it].salary))

for it,row in  data_fr.iterrows():
    if type(row.job_name) is float:
        continue
    dic_crawled_job_to_insee.setdefault(unidecode(row.job_name.lower()),
                                        (data_fr.iloc[it].insee_job,
                                        data_fr.iloc[it].salary))

dic_cat={}
for k,v in dic_crawled_job_to_insee.items():
    mat=re.search("[0-9]{3}[a-z]",v[0])
    if  mat.group(0):
        alterna=re.sub("(\([a-z0-9,. - ]+\)(.*?))","",k)
        dic_cat.setdefault(mat.group(0),[])
        dic_cat[mat.group(0)].append(alterna)
        all_prods.append(alterna)
        
tw_pr=[]
f=open("/warehouse/COMPLEXNET/jlevyabi/ml_soc_econ/icdm18/issues/profile_to_test.txt","r")
for line in tqdm(f.readlines()):
    tw_pr.append(unidecode(line.lower()[:-2]))


```

```python
import pandas as pd
data_jobs=pd.read_csv("/warehouse/COMPLEXNET/jlevyabi/ml_soc_econ/data_files/expicit_cat.csv",sep=";")
```

```python
#### from tqdm import tqdm_notebook as tqdmn
from fuzzywuzzy import fuzz
twitter_geol_profiles=sorted(list(set([" ".join(x.split()) for x in tw_pr])))
jobs=set(data_jobs.Specific)

susp=[]
for ind_pr in tqdmn(twitter_geol_profiles):
    res=ind_pr
    boo=False
    for job in jobs:
        if type(job) is float or type(ind_pr) is float:continue
        job=re.sub(r'[\W_]+', ' ', job)
        ind_pr=re.sub(r'[\W_]+', ' ',ind_pr)
        job_arr,prof_arr=job.split(),ind_pr.split()
        fuzz_val=fuzz.ratio(ind_pr, job)
        overlap_val=0
        if len(job_arr)>0 and len(prof_arr)>0:
            overlap_val=len(set(job_arr).intersection(set(prof_arr)))/(min(len(job_arr),len(prof_arr))+0.0)        
        if fuzz_val >70:
            res+=(" "+'\x1b[6;30;42m'+job+"/"+'\x1b[0m')
            boo=True
        if overlap_val>0.5:
            res+=(" "+'\033[94m'+job+"/"+'\x1b[0m')
            boo=True
    if boo:
        print(res)
        
            
```
